```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r'C:\Users\SG\Downloads\xgtutorial.csv')
```

```python
df
```

```python
a_xG =[0]
h_xG =[0]
a_min =[0]
h_min =[0]

hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]

for x in range(len(df['xG'])):
    if df['team'][x] ==ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x] ==hteam:
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
a_xG
```

```python
def cum(num_list):
    return [sum(num_list[:i+1]) for i in range(len(num_list))]

a_cum=cum(a_xG);
h_cum=cum(h_xG);
```

```python
a_cum
h_cum
```

```python
fig, ax= plt.subplots(figsize=(10,5)) 
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Minutes')
plt.ylabel('xG')

ax.step(x=a_min, y=a_cum)
ax.step(x=h_min, y=h_cum)
```
