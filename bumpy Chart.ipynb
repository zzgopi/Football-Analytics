```python
import pandas as pd
import matplotlib.pyplot as plt
import mplsoccer
import highlight_text
from mplsoccer import Bumpy
import numpy as np

from PIL import Image
from io import BytesIO
import requests
```

```python
df=pd.read_csv('la liga.csv')
```

```python
df=df.T
```

```python
df
```

```python
df.columns=df.iloc[0]
df=df.iloc[1:]
df
```

```python
week = ['Week ' + str(num) for num in range(1,35)]
high = {
    'Atletico Madrid':'#CB3524',
    'Real Madrid' : 'white',
    'Barcelona':'#EDBB00'
}
```

```python
bumby =Bumpy(
    scatter_color= '#282828',line_color='#252525',
    rotate_xticks=90,
    ticklabel_size=15,
    scatter_primary='D',
    show_right=True,
    plot_labels =True,
    alignment_yvalue=.1,
    alignment_xvalue=.065
)
```

```python
fig, ax= bumby.plot(
    x_list=week,
    y_list=np.linspace(1,20,20).astype(int),
    values=df,
    secondary_alpha=.5,
    highlight_dict=high,
    figsize=(18,18),
    y_label= 'Postion',
    ylim=(-2,23),
    lw=2.5
)

fig.text(s='La Liga through Week 34',x=.35,y=.85,
         c='white',size=24,weight='bold' )

highlight_text.fig_text(x=.35,y=.82,
                       s='Comparing <Atletico Madrid>,<Real Madrid> and <Barcelona> ',
                       highlight_textprops = [
                           {"color":'#CB3524'},
                           {"color":'white'},
                           {"color":'#EDBB00'}  
                       ],
                        fontsize=20,
                        color='Blue'
                       )
ax2=fig.add_axes([.15,.79,.13,.13])
ax2.axis('off')
img=Image.open('laligalogo.png')
ax2.imshow(img)
```
