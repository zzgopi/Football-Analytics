```python
import pandas as pd
from mplsoccer.pitch import Pitch
```

```python
df= pd.read_csv(r'C:\Users\SG\Downloads\valladolidA.csv')
```

```python
df
```

```python
df=df[df['teamId']=='Barcelona']
```

```python
df
```

```python
df['passer']= df['playerId']
df['recipient'] = df['playerId'].shift(-1)

passes= df[df['type']=='Pass']
succesful= passes[passes['outcome']=='Successful']
```

```python
succesful
```

```python
subs=df[df['type']=='SubstitutionOff']
subs= subs['minute']
firstSub=subs.min()
```

```python
succesful= succesful[succesful['minute']<firstSub]
succesful
```

```python
pas = pd.to_numeric(succesful['passer'],downcast='integer')
rec = pd.to_numeric(succesful['recipient'],downcast='integer')

succesful['passer']= pas
succesful['recipient'] = rec

succesful


```

```python
avg_location   = succesful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
avg_location.columns =['x','y','count']
```

```python
avg_location
```

```python
pas_between= succesful.groupby(['passer','recipient']).id.count().reset_index()
pas_between.rename({'id':'pass_count'},axis='columns',inplace=True)

pas_between = pas_between.merge(avg_location, left_on='passer', right_index= True)
pas_between = pas_between.merge(avg_location, left_on='recipient', right_index= True,suffixes=['','_end'])
```

```python
pas_between
```

```python
pas_between = pas_between[pas_between['pass_count']>3]
```

```python
pas_between
```

```python
pitch=Pitch(pitch_type='statsbomb', orientation='horizontal', pitch_color='#22312b', line_color='#c7d5cc', figsize=(16,11), constrained_layout=False, tight_layout=True)
fig, ax = pitch.draw()

arrows= pitch.arrows(1.2*pas_between.x, .8*pas_between.y,1.2* pas_between.x_end,.8*pas_between.y_end,ax=ax,
                    width=3,headwidth=3, color='blue', zorder=1, alpha =.5)
nodes =pitch.scatter(1.2*avg_location.x,.8*avg_location.y,s=300,color='#d3d3d3',  zorder=1, alpha =.5, ax=ax)
```
