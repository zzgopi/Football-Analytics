```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sms
```

```python
df= pd.read_csv(r'C:\Users\SG\Downloads\beeswarmTutorial.csv')

df['per90']=df['Prog']/df['90s']

df
```

```python
df = df[df['90s']>=6.5]
df
```

```python
df =df[df['Pos']!= 'GK'].reset_index()
```

```python
df
```

```python
backgd= '#313332'
text_color= 'white'

fig, ax= plt.subplots(figsize=(10,5))
fig.set_facecolor(backgd)
ax.patch.set_facecolor(backgd)

mpl.rcParams['xtick.color']=text_color
mpl.rcParams['ytick.color']=text_color

spines =['top','bottom','left','right']
for x in spines:
        ax.spines[x].set_visible(False)

sms.swarmplot(x='per90',data=df,zorder=1)

plt.scatter(x=6.495726,y=0,c='red',s=200,zorder=2)
```

```python
df2=pd.read_csv(r'C:\Users\SG\Downloads\beeswarm2.csv')
```

```python
df2
```

```python
df2['Player']=df2['Player'].str.split('\\',expand=True)[0]
df2 = df2[df2['90s']>=6.5]
df2 =df2[df2['Pos']!= 'GK'].reset_index()
df2
```

```python
df2['1/3']=df2['1/3']/df2['90s']
df2['PPA']=df2['PPA']/df2['90s']
df2['xA']=df2['xA']/df2['90s']
df2['KP']=df2['KP']/df2['90s']
df2['Prog']=df2['Prog']/df2['90s']
```

```python
df2
```

```python
metrics = ['Prog','1/3','xA','Cmp%','KP','PPA']
```

```python
fig,axes = plt.subplots(3,2,figsize=(14,10))
fig.set_facecolor(backgd)
ax.patch.set_facecolor(backgd)

#set up our base layer
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color

#create a list of comparisons
counter=0
counter2=0
met_counter = 0

for i,ax in zip(df2['Player'],axes.flatten()):
    ax.set_facecolor(backgd)
    ax.grid(ls='dotted',lw=.5,color='lightgrey',axis='y',zorder=1)
    
    spines = ['top','bottom','left','right']
    for x in spines:
        if x in spines:
            ax.spines[x].set_visible(False)
            
    sms.swarmplot(x=metrics[met_counter],data=df2,ax=axes[int(met_counter/2),int(met_counter%2)],zorder=1,color='#64645e')
    ax.set_xlabel(f'{metrics[met_counter]}',c='white')
    
    for x in range(len(df2['Player'])):
        if df2['Player'][x] == 'Jack Grealish':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,c='white',zorder=2)
    
    for x in range(len(df2['Player'])):
        if df2['Player'][x] == 'Wilfried Zaha':
            ax.scatter(x=df2[metrics[met_counter]][x],y=0,s=200,c='red',zorder=2)
                        
    met_counter+=1
```
